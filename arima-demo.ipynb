## Demo of an ARIMA model for time-series forecasting of backup repositories

Demonstration of how to build and run an ARIMA(p,d,q) model for capacity planning of a backup repositories. The dataset contains real-world capacity data of some Veeam Backup Repositories. <br><br>Check out the *vcap* Capacity Reporter Script [here](https://github.com/StefanSpecht/vcap) to see the model in action! <br> A demo how to grid-search the ARIMA hyper-parameters can be found [here](https://github.com/StefanSpecht/arima-demo/blob/master/arima-gridsearch.jpynb). <br> <br>
Demo is inspired by the following tutorial: https://machinelearningmastery.com/arima-for-time-series-forecasting-with-python/

First, we load the dataset and bring it in shape a little bit

```python
%matplotlib inline

import pandas as pd
from matplotlib import pyplot
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
from datetime import datetime, timedelta

df = pd.read_csv('repository_data.csv', header=0, index_col=0, parse_dates=True, dayfirst=True)
df['UsedGB'] = df['StorageTotal'] - df['StorageFree']
df = df[['Name', 'UsedGB', 'StorageTotal']]
```

Then, we filter the dataframe for the repository of interest

```python
df = df[df['Name'] == "ES1_EE2203_Backup_01"]
print(df.head())
```

Some dates migt be missing within the series, so let's re-index series and forward-fill missing values

```python
series = pd.Series(df['UsedGB'])
date_range = pd.date_range(series.index[1].date(), series.index[-1].date())
series = series.reindex(date_range)
series.fillna(method='ffill', inplace=True)
```

Next, let's visualize the series by ploting it

```python
series.plot()
pyplot.show()
```

For building our ARIMA(p,d,q) model we need to determine the three hyper-parameters:
 - p: number of lags for the Auto-regression model **(AR)**
 - d: degree of differencing the series to make it stationary if needed **(I)**
 - q: moving average window size **(MA)**

A clear (upwards) trend can be seen in the plot which indicates that the series is not stationary. To make the series stationary it must be differenced once. That will be done by specifiying an ARIMA(p,d,q) parameter of d=1

Let's analze data for autocorrelation

```python
pd.plotting.autocorrelation_plot(series)
pyplot.show()
```

There is significant positive autocorrelation for the first about 10 lags. To not build a model that is too computational expensive let's set the p-parameter to 5.

Time to create the ARIMA model now and fit it

```python
# fit model
model = ARIMA(series, order=(5,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

Let analyze the observed residuals errors by plotting them and also check the distribution of the density plot:

```python
# plot residual errors
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()
residuals.plot(kind='kde')
pyplot.show()
print(residuals.describe())
```

Seems like all trends were captured by our model and the density of residual errors is Guassian distributed and quite nicely centered around zero. 

We split our series now into a training and a testing set and run our model to validate it's performance

```python
X = series.values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
	model = ARIMA(history, order=(5,1,0))
	model_fit = model.fit(disp=0)
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = test[t]
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```

Our model did a quite good job so far. Finally let's calculate, when we reach 100% capacity on our backup repository

```python
history = [x for x in series.values]
hard_threshold = df['StorageTotal'].iloc[-1]
soft_threshold = hard_threshold * 0.9
current_date = series.tail(1).index.date
#predictions = list()
predicted_series = pd.Series()
i=0

while True:
    i = i+1
    current_date += timedelta(days=1)
    model = ARIMA(history, order=(6,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast(steps=1)
    yhat = output[0]
    #print (yhat)
    print('predicted=%f, date=%s' % (yhat, current_date))
    predicted_series = predicted_series.append(pd.Series(yhat, index=current_date))
    history.append(yhat[0])
    if (yhat >= hard_threshold):
        break
```

```python
#scale
if hard_threshold > 1024:
    series = series/1024
    predicted_series = predicted_series/1024
    hard_threshold = hard_threshold/1024
    soft_threshold = soft_threshold/1024
    ylabel = 'Used [TB]'
else:
    ylabel = 'Used [GB]'
#Plot the data
fig = pyplot.figure()
pyplot.title('Sample Repository')
pyplot.plot(series)
pyplot.plot(predicted_series, color='red')
pyplot.axhline(y=hard_threshold, color='r')
pyplot.axhline(y=soft_threshold, color='y', ls=':')
pyplot.ylabel(ylabel)
```
